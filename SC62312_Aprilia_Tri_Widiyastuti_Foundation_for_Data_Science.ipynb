5.	**Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.**
*  Normal: 18.5-24.9
*   Overweight: 25.0-29.9
*   Obesity: 30.0-34.9
*   Underweight: < 18.5

```python
def status_bmi(bmi):
  if bmi < 18.5:
    return ('Underweight')
  elif 18.5 <= bmi <= 24.9:
    return ('Normal')
  elif 25.0 <= bmi <= 29.9:
    return ('Overweight')
  elif 30.0 <= bmi <= 34.9:
    return ('Obesity')
  else:
    return ('Obesity Berlebih')

data['Status BMI'] = data['BMI'].apply(status_bmi)
print(data)
```

SOAL 2
Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

Contoh:
Input
>> Money Capital : 100000
>> Investment Tenure (years) : 2
Output
>> Year 1, your total investment : 125000.0
>> Year 2, your total investment : 156250.0


```python
modal_invest = int(input("Masukan nominal: "))
jangka_invest = int(input("Masukan tahun: "))

def investasi(modal, jangka_waktu):
  if modal < 0 or modal > 1000000000:
    print("Error: Modal atau tenure tidak sesuai ketentuan.")
    return

  if jangka_waktu < 1 or jangka_waktu > 25:
    print("Error: modal atau tenure tidak sesuai ketentuan.")
    return

  for i in range(1, jangka_waktu + 1):
    modal *= 1.25

    print(f"Tahun {i}, total investasi : {modal}")

investasi(modal_invest, jangka_invest)
```
